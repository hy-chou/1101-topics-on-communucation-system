## QSVM circuit

```python
def qsvm_circuit(train,theta,dot_val):
    q = QuantumRegister(5, name="x")
    c = ClassicalRegister(2, name="c")
    qc = QuantumCircuit(q, c)
    qc.reset(q)

    matrix = np.array([[1.05, dot_val], [dot_val, 1.05]])
    vector = np.array([1, 0])
    hhl = HHL().construct_circuit(matrix, vector)
    qc.compose(hhl, qubits=[2,1,0,3], inplace=True)

    qc.barrier()

    qc.x(q[2])
    qc.ccx(q[2], q[3], q[4])
    qc.x(q[2])
    qc.cry(train[0], q[4], q[1])
    qc.x(q[2])
    qc.ccx(q[2], q[3], q[4])
    qc.x(q[2])

    qc.barrier()

    qc.ccx(q[2], q[3], q[4])
    qc.cry(train[1], q[4], q[1])
    qc.ccx(q[2], q[3], q[4])

    qc.barrier()

    qc.cry(-theta, q[3], q[1])
    qc.ch(q[3], q[2])

    qc.barrier()

    qc.h(q[3])
    qc.measure(q[3], c[0])
    qc.h(q[1])
    qc.measure(q[1], c[1])

    qc.draw(output='mpl')
    return qc
```

```python
def qsvm_circuit2(train,theta,dot_val):
    q = QuantumRegister(5, name="x")
    c = ClassicalRegister(2, name="c")
    qc = QuantumCircuit(q, c)

    qc.u(-1.57, 0, 0, q[2])    # qc.u3(-1.57, 0, 0, q[2])

    qc.barrier()

    qc.h(q[0])
    qc.h(q[1])

    qc.barrier()

    qc.p(3.17, q[0])    # qc.u1(3.17, q[0])
    qc.p(1.59, q[1])    # qc.u1(1.59, q[1])
    qc.cu(2*dot_val*pi, pi/2, -pi/2, 0, q[0], q[2])    # qc.cu3(pi, pi/2, -pi/2, q[0], q[2])
    qc.cu(dot_val*pi, pi/2, -pi/2, 0, q[1], q[2])    # qc.cu3(pi/2, pi/2, -pi/2, q[1], q[2])

    qc.barrier()

    qc.h(q[1])
    qc.cp(-pi/2, q[0], q[1])    # qc.cu1(-pi/2, q[0], q[1])
    qc.h(q[0])

    qc.barrier()

    qc.cry(pi/4, q[0], q[3])
    qc.cry(pi/8, q[1], q[3])

    qc.barrier()

    qc.h(q[0])
    qc.cp(pi/2, q[0], q[1])    # qc.cu1(pi/2, q[0], q[1])
    qc.h(q[1])

    qc.barrier()

    qc.p(-1.59, q[0])
    qc.p(-3.17, q[1])
    qc.cu(-dot_val*pi, pi/2, -pi/2, 0, q[0], q[2])    # qc.cu3(-pi/2, pi/2, -pi/2, q[0], q[2])
    qc.cu(-2*dot_val*pi, pi/2, -pi/2, 0, q[1], q[2])    # qc.cu3(-pi, pi/2, -pi/2, q[1], q[2])

    qc.barrier()

    qc.h(q[0])
    qc.h(q[1])

    qc.barrier()

    qc.x(q[2])
    qc.ccx(q[2], q[3], q[4])
    qc.x(q[2])
    qc.cry(train[0], q[4], q[1])
    qc.x(q[2])
    qc.ccx(q[2], q[3], q[4])
    qc.x(q[2])

    qc.barrier()

    qc.ccx(q[2], q[3], q[4])
    qc.cry(train[1], q[4], q[1])
    qc.ccx(q[2], q[3], q[4])

    qc.barrier()

    qc.cry(-theta, q[3], q[1])
    qc.ch(q[3], q[2])

    qc.barrier()

    qc.h(q[3])
    qc.measure(q[3], c[0])
    qc.h(q[0])
    qc.measure(q[0], c[1])
    
    return qc
```
